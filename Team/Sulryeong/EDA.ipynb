```python
bins = [i for i in range(0, 6801, 100)]
bins2 = [i for i in range(0, 6801, 10)]
```

```python
order_items['price_bins'] = pd.cut(order_items.price, bins, right=True, labels=bins[1:])
order_items['price_bins2'] = pd.cut(order_items.price, bins2, right=True, labels=bins2[1:])
```

```python
order_items.head()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot(data=order_items, x='price_bins')

plt.xticks(rotation=90)
plt.title('제품 가격 분포(~이하)')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot(data=order_items[order_items.price_bins==100], x='price_bins2', order=[i for i in range(0, 101, 10)][1:])

plt.xticks(rotation=90)
plt.title('제품 가격 분포(100 이하)')
plt.show()
```

#### 가설 확인

```python
late_delivered_id = orders[orders.order_delivered_customer_date > orders.order_estimated_delivery_date].order_id.to_list()
```

```python
late_del_review_sc = order_reviews[order_reviews.order_id.isin(late_delivered_id)].review_score.value_counts()
late_del_review_sc
```

```python
labels = late_del_review_sc.index.to_list()
explodes = [0.05, 0.05, 0.05, 0.05, 0.05]

fig = plt.figure(figsize=(5, 5))
plt.pie(late_del_review_sc, labels=labels, autopct='%1.2f%%', explode=explodes)
plt.title('리뷰 점수 비율(%) - 예상 도착일보다 실제 배송 도착일이 늦은 경우')
plt.show()
```
