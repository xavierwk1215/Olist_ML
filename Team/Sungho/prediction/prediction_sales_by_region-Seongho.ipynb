```python
# 필요한 열 선택
columns = ['order_purchase_timestamp', 'customer_state', 'payment_value']
df = olist[columns]
df
```

```python
# 특정 customer_state 선택
selected_state = 'AC'  # 원하는 주(state) 입력

# 특정 주(state)에 해당하는 데이터 추출
state_data = df_grouped[df_grouped['customer_state'] == selected_state]

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(state_data['order_purchase_timestamp'], state_data['payment_value'])
plt.title(f'Daily Sales in {selected_state}')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler


# 'payment_value' 열의 값만 추출
data = df['payment_value'].values.astype(float)

# 데이터 정규화 (0과 1 사이로 스케일링)
scaler = MinMaxScaler()
data = scaler.fit_transform(data.reshape(-1, 1))

# 데이터 윈도우 설정
def create_windowed_data(data, window_size):
    X, Y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        Y.append(data[i+window_size])
    return np.array(X), np.array(Y)

window_size = 10
X, Y = create_windowed_data(data, window_size)
```

```python
# LSTM 모델 구성
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(window_size, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# 모델 학습
model.fit(X, Y, epochs=50, batch_size=32)
```

```python
# 예측
future_window = data[-window_size:].reshape(1, window_size, 1)
forecast_steps = 30  # 예측할 단계 수 조정 (예: 30일)
forecast = []

for _ in range(forecast_steps):
    next_step = model.predict(future_window)
    forecast.append(next_step[0, 0])
    future_window = np.roll(future_window, shift=-1, axis=1)
    future_window[0, -1, 0] = next_step[0, 0]

# 역 스케일링 (정규화 해제)
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# 결과 시각화
plt.plot(df.index, df['payment_value'], label='실제 데이터')
plt.plot(df.index[-1] + np.arange(1, forecast_steps+1), forecast, marker='o', markersize=5, label='예측')
plt.legend()
plt.title('시계열 예측')
plt.show()
```
